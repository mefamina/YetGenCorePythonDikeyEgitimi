# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız

class Vehicle():
    def __init__(self, marka, model, silindir, renk):
        self.marka = marka
        self.model = model
        self.silindir = silindir
        self.renk = renk
    def intro(self):
        print(self.marka, self.model,self.silindir, self.renk)

class Bus(Vehicle):
    def __init__(self, marka, model, silindir, renk, kapasite):
        super(). __init__(marka, model, silindir, renk)
        self.kapasite = kapasite
    def intro(self):
        print("Bus nesnesi oluşturuldu")
    def bilgiler(self):
        print(f"""
        ****Otobüs Bilgileri****
        Marka: {self.marka}
        Model: {self.model}
        Silindir: {self.silindir}
        Renk: {self.renk}
        Kapasite: {self.kapasite}""")

otobüs = Bus("MAN", "Z-449", 4, "Siyah", 50)
otobüs.intro
otobüs.bilgiler()
````
